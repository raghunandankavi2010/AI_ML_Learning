Create a pivot table with the mother’s education in rows and the number of siblings in columns and the values (or vice versa).

```python
df['Siblings'] = df.Siblings.replace('Single child','0')
```

```python
def extract_number(val):
    return int(val[0])     

df['Siblings_count'] = df.Siblings.apply(extract_number)
```

```python
df.head()
```

```python
# to find frequencey or count use len as aggfunc in pivot table
pivot_df = pd.pivot_table(df, index="Siblings", columns=['Mother.edu'], values="Siblings_count",aggfunc=len) 

pivot_df
```
