```python
odi_df = pd.read_csv('odi-batting.csv')
odi_df.head(10)
```

```python
odi_df['Century'] = odi_df['Runs'].apply(lambda x: 1 if x >= 100 else 0)
```

```python
odi_df.head(10)
```

```python
century_counts = odi_df.groupby('Player')['Century'].sum()
century_counts.sort_values(ascending=False, inplace=True)
```

```python
century_counts.head()
```

```python
odi_df['Strike Rate'] = odi_df['Runs'] / odi_df['Balls'] * 100
```

```python
odi_df.head(10)
```

```python
pivot_df = pd.pivot_table(odi_df, index="Player", columns=['Century'], values="Strike Rate",aggfunc=np.max)
```

```python
player_stats = pivot_df.loc[['Shahid Afridi', 'Mark V Boucher', 'Sanath T Jayasuriya','Shane R Watson']] 

player_stats
```

```python
indian_players = odi_df[odi_df['Country'] == "India"]
indian_players.head()
```

```python
indian_players['year'] = pd.to_datetime(indian_players['MatchDate']).dt.year
indian_players.head()
```

```python
centuries_by_year = indian_players.groupby('year')['Century'].sum()

# get the year with highest number of centuries
year_with_highest_centuries = centuries_by_year.idxmax()
year_with_highest_centuries
```

Extracting Day From a Date
Description
The given dataset 'order' has a variable 'Order_Date' with the dates of purchase.
Create a new variable 'day' which will contain the day from the date at variable Order_Date

```python
order = pd.read_csv('https://query.data.world/s/3hIAtsCE7vYkPEL-O5DyWJAeS5Af-7')
order['Order_Date'] = pd.to_datetime(order['Order_Date'])

order.head()

```

```python
order['day'] = order['Order_Date'].dt.day
```
