```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("E:\\AI_ML_Learning\\EDA\\tendulkar_ODI.csv")
df.head()
```

```python
df.shape
```

```python
df.Mins.dtype
```

```python
# clean column mins
def clean_mins(val):
    return int(val.replace('-','0'))

df['Mins'] = df.Mins.apply(clean_mins)
```

```python
df.head()
```

```python
# rename Unamed:0 column
df.rename(columns={'Unnamed: 0': 'Rank'}, inplace=True)
```

```python
df.head()
```

```python
# clean column 4s
df_with_dash = df.loc[df['4s'] == '-']
df_with_dash
```

```python
df['Runs'] = df.Runs.apply(lambda x: x.replace('\*',''))
```

```python
df['Runs'] = df.Runs.apply(lambda x: x.replace('DNB','0'))
```

```python
df['Runs'] = df.Runs.apply(lambda x: x.replace('TDNB','0'))
```

```python
df['4s'] = df['4s'].replace('-', 0)
```

```python
df['4s'] = df['4s'].replace('-', 0)
```

```python
df['SR'] = df['SR'].replace('-', 0)
```

```python
df['Dismissal'] = df['Dismissal'].replace('-', 'not dismissed')
```

```python
df['Pos'] = df['Pos'].replace('-', 0)
```

```python
df_with_dash = df.loc[df['4s'] == '-']
df_with_dash
```

```python
df.head(30)
```

```python
df['Runs'] = df['Runs'].str.replace('*', '')
```

```python
def clean_runs(val):
    return int(val.replace('*',''))

res= clean_runs("82*")
res
df['Runs'] = df.Runs.apply(lambda x: x.replace('T0','0'))
```

```python
df['Runs'] = df['Runs'].astype('int')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['Runs'].value_counts().sort_index().plot(kind='bar')

# set the x-label and y-label
plt.xlabel('Runs')
plt.ylabel('Frequency')

# show the plot
plt.show()
```

```python
def convert_string(val):
    return str(val)

df['4s'] = df.Mins.apply(convert_string)
```

```python
plt.figure(figsize = (18 , 5), dpi = 100)
sns.distplot(df['4s'] , bins = 20, color = "teal")
plt.xticks(rotation = 90);
plt.show()
```
