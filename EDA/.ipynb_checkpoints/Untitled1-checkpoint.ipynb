```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("E:\\AI_ML_Learning\\EDA\\tendulkar_ODI.csv")
df.head()
```

```python
df.shape
```

```python
df.Mins.dtype
```

```python
# clean column mins
def clean_mins(val):
    return int(val.replace('-','0'))

df['Mins'] = df.Mins.apply(clean_mins)
```

```python
df.head()
```

```python
# rename Unamed:0 column
df.rename(columns={'Unnamed: 0': 'Rank'}, inplace=True)
```

```python
df.head()
```

```python
# clean column 4s
df_with_dash = df.loc[df['4s'] == '-']
df_with_dash
```

```python
df['4s'] = df['4s'].replace('-', 0)
```

```python
df['4s'] = df['4s'].replace('-', 0)
```

```python
df['SR'] = df['SR'].replace('-', 0)
```

```python
df['Dismissal'] = df['Dismissal'].replace('-', 'not dismissed')
```

```python
df['Pos'] = df['Pos'].replace('-', 0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['Runs'].value_counts().sort_index().plot(kind='bar')

# set the x-label and y-label
plt.xlabel('Runs')
plt.ylabel('Frequency')

# show the plot
plt.show()
```

```python
def convert_string(val):
    return str(val)

df['4s'] = df.Mins.apply(convert_string)
```

```python
sns.displot(df, x="4s",binwidth=10,bins=15)
plt.xticks(rotation="vertical")

plt.xticks(df['4s'][::20])
plt.show()
```
