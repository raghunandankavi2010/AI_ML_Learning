Create a pivot table with the motherâ€™s education in rows and the number of siblings in columns and the values (or vice versa).

```python
df['Siblings'] = df.Siblings.replace('Single child','0')
```

```python
def extract_number(val):
    return int(val[0])     

df['Siblings_count'] = df.Siblings.apply(extract_number)
```

```python
df.head()
```

```python
# to find frequencey or count use len as aggfunc in pivot table
pivot_df = pd.pivot_table(df, index="Siblings", columns=['Mother.edu'], values="Siblings_count",aggfunc=len) 

pivot_df
```

```python
df['Science..'] = df['Science..'].fillna(0)
```

```python
# to find frequencey or count use len as aggfunc in pivot table
#Compare the science percentage in a pivot table with the fatherâ€™s education and age in rows and columns (in any order).
pivot_df2 = pd.pivot_table(df, index="Father.edu", columns=['Age'], values="Science..",aggfunc=lambda x: sum(x) / sum(df['Science..']) * 100) 

pivot_df2 
```

You will see that children whose fathers have a degree and above perform better in science only in the age groups 12, 13 and 14 years.
